```python
import pandas as pd
import numpy as np
```

```python
# import rfc_dat_avg.csv
rfc_dat_avg = pd.read_csv('rfc_dat_avg.csv')
rfc_dat_avg.head()
# set each row name equal to the income_bracket
rfc_dat_avg.index = rfc_dat_avg['income_bracket']
# remove unnamed column
rfc_dat_avg = rfc_dat_avg.drop('Unnamed: 0', axis=1)
rfc_dat_avg = rfc_dat_avg.drop('income_bracket', axis=1)
rfc_dat_avg.head()
```

```python
# get the top 10 features for each income bracket
very_low_10 = rfc_dat_avg.loc['very low'].sort_values(ascending=False).head(10)
very_low_10

low_10 = rfc_dat_avg.loc['low'].sort_values(ascending=False).head(10)
low_10

medium_10 = rfc_dat_avg.loc['medium'].sort_values(ascending=False).head(10)
medium_10

high_10 = rfc_dat_avg.loc['high'].sort_values(ascending=False).head(10)
high_10

very_high_10 = rfc_dat_avg.loc['very high'].sort_values(ascending=False).head(10)
very_high_10
```

```python
# merge each income bracket's top 10 features into one dataframe
top_10 = pd.concat([very_low_10, low_10, medium_10, high_10, very_high_10], axis=1)
print(top_10)
```

```python
# T010101: Sleeping
# T050101: Working, main job
# T120303: Television and movies (not religious)
# T110101: Eating and drinking
# T010201: Washing, dressing, and grooming oneself
# T120101: Socializing and communicatins with others
# T020101 Interior cleaning
# T020201: Food and drink preparation
# T120301: Relaxing, thinking
# T180501: Travel related to working
# T070104: Shopping, except groceries, food and gas
# Rename the rows in the pandas dataset
top_10 = top_10.rename(index={'T010101': 'Sleeping', 'T050101': 'Working, main job', 'T120303': 'Television and movies (not religious)', 'T110101': 'Eating and drinking', 'T010201': 'Washing, dressing, and grooming oneself', 'T120101': 'Socializing and communicatins with others', 'T020101': 'Interior cleaning', 'T020201': 'Food and drink preparation', 'T120301': 'Relaxing, thinking', 'T180501': 'Travel related to working', 'T070104': 'Shopping, except groceries, food and gas'})
top_10
```
